```python
import pandas as pd
```

```python
path='C://Users/sergi/Documents/GitHub/py-villacres-lizano-sergio-wladimir/Examen/articulos.csv'
```

```python
lista = pd.read_csv(
        path
        )
lista
```

```python
nueva_lista =lista.Productos.apply(lambda x: pd.Series(x.split(' ')))
nueva_lista
```

```python
data = pd.DataFrame(columns=('grupo', 'subgrupo'))
grupos=[]
subgrupos=[]
```

```python
for item, fila in nueva_lista.iterrows():
    if(pd.isnull(fila[1])):
        grupos.append(fila[0])
        subgrupos.append(fila[0])
    elif(pd.isnull(fila[2])):
        grupos.append(fila[0]+' '+fila[1])
        subgrupos.append(fila[0]+' '+fila[1])
    else:
        grupos.append(fila[0])
        valor = len(fila)
        sub=''
        for it in range(2,valor-1,1):
            sub=sub+' '+str(fila[it])
        subgrupos.append(sub)
    
```

```python
serie_grupo = pd.Series(grupos)
serie_subgrupo = pd.Series(subgrupos)
```

```python
data['grupo']=serie_grupo
data['subgrupo']=serie_subgrupo
data
```

```python
data.to_csv('archivo_fin.csv', header=True, index=False)
```
