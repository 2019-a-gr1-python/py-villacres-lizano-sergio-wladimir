```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path='C://Users/sergi/Documents/GitHub/py-villacres-lizano-sergio-wladimir/Proyectos/Segundo Bimestre/scrapy_proyecto/tmp/productos-mercado-libre.csv'
```

```python
lista = pd.read_csv(
        path
        )
```

```python
lista
```

```python
p = sns.barplot(x='titulo', y='vendidos', data=lista)
```

```python
data_gdp = (lista['precio'].groupby(lista.lugar)).sum()
data_gdp.plot(figsize=(50,20),fontsize=50,color='red')
plt.xlabel('Lugar', fontsize=50)
plt.ylabel('Precio',fontsize=50)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(15, 10))
ax = sns.regplot(x='precio', y='vendidos',data=lista)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(lista.groupby(['titulo']).vendidos.sum(),'ko--')
ax.set_title('Ventas por marca')
ax.set_xlabel('Marca')
ax.set_ylabel('Vendidos')
```

```python
data_suicide_mean = lista['precio'].groupby(lista.titulo).mean().sort_values(ascending=False)
f,ax = plt.subplots(1,1,figsize=(15,4))
ax = sns.barplot(data_suicide_mean.head(10).index,data_suicide_mean.head(10),palette='coolwarm')
```

```python
g=sns.lineplot(x='lugar', y='vendidos',hue=lista.titulo,data=lista)
```
