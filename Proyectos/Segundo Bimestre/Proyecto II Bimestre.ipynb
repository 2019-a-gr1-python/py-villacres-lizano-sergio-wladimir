##### Determinar que marca de telefono se vende mas

```python
p = sns.barplot(x='titulo', y='vendidos', data=lista)
```

Analisis: Como se puede observar en el grafico la marca de telefonos que mas se vende es Huawei seguido de Xiaomi y despues samsung, mientras que los que menos se venden son Motorola y Sony Xperia. Esto se puede deber a los precios mas accesibles en estas marcas lo cual comprobaremos mas adelante.

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(1, 1, 1)
ax.plot(lista.groupby(['titulo']).vendidos.sum(),'ko--')
ax.set_title('Ventas por marca')
ax.set_xlabel('Marca')
ax.set_ylabel('Vendidos')
```

Al realizar este nuevo grafico observamos que la marca mas vendida es Samsung seguida de Xiaomi y despues Huawei, la diferencia con la grafica anterior se debe a que esta vez agrupamos los datos y sumamos por marca, lo cual la anterior grafica no representa, en este caso la primera grafica representa el articulo que en una sola publicacion vendio mas por marca.

##### Determinar que sector es el que mas compra en mercado libre

```python
data_gdp = (lista['precio'].groupby(lista['lugar'])).sum()
data_gdp.plot(figsize=(50,20),fontsize=50,color='blue')
plt.xlabel('Lugar', fontsize=50)
plt.ylabel('Precio',fontsize=50)
plt.show()
```

Analisis: Este grafico no nos muesta todos los lugares que tenemos en los datos, pero esto se puede deber a que los lugares son tantos que el eje x no puede llenarce completamente

```python
data = lista['vendidos'].groupby(lista.lugar).mean().sort_values(ascending=False)
f,ax = plt.subplots(1,1,figsize=(15,4))
ax = sns.barplot(data.head(10).index,data.head(10),palette='coolwarm')
```

Como podemos observar al obtener el top 10 de los lugares que mas telefonos se venden obtenemos que Quito es la ciudad que mas comercializa telefonos en esta plataforma, esto nos puede ayudar a concluir que el mercado al que va enfocado mercado libre es a Quito.

```python
data_suicide_mean = lista['precio'].groupby(lista.lugar).mean().sort_values(ascending=False)
f,ax = plt.subplots(1,1,figsize=(15,4))
ax = sns.barplot(data_suicide_mean.head(10).index,data_suicide_mean.head(10),palette='coolwarm')
```

Analisis: En esta grafica vemos representado el lugar que mas costoso vende los dispositivos, y esta puede ser una razon importante para analizar el comportamiento de las ventas en relacion al precio.
