## 28.- Modificar la imagen, con rgb(0,0,255),rgb(0,255,0),rgb(255,0,0) 

```python
mapache = misc.face()
```

```python
def muestra_imagen(imagen):
    plt.imshow(imagen)
    plt.show()
    
muestra_imagen(mapache)
```

```python
mapache_azul = np.copy(mapache)
mapache_azul[:,:,0] = 0
mapache_azul[:,:,1] = 0
muestra_imagen(mapache_azul)
```

```python
mapache_verde = np.copy(mapache)
mapache_verde[:,:,0] = 0
mapache_verde[:,:,2] = 0
muestra_imagen(mapache_verde)
```

```python
mapache_rojo = np.copy(mapache)
mapache_rojo[:,:,1] = 0
mapache_rojo[:,:,2] = 0
muestra_imagen(mapache_rojo)
```
